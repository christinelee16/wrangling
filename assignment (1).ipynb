1. Data tidying helps make it easier to clean and visualize data. Tidy datasets require a small set of tools to deal with a wide range of un-tidy data.
2. The "tidy data standard" aims to provide a clear and consistent way of organizing and structuring data to make it easier to clean and work with. The tidy data standard ensures data is organized in a way that is intuitive to analysis and visualization. The key goals of the tidy data standard are consistency, clarity, and simplicity. In a tidy dataset, each variable forms a column and each observation forms a row.
3. Tidy datasets all follow a common structure, the way families follow a common structure. However, messy datasets don't follow a consistent structure, and they can be disorganized in unique ways.
When given a dataset, identifying which elements are observations or variables is usually straightforward. However, defining what is a variable or an observation can be complicated.
4. According to Wickham, values are usually either numbers or strings. Variables contain vlaues that measure the same attribute across units. An observation contain values that are measured on the same unit across attributes.
5. In tidy data, each variable forms a column, each observation forms a row, and each type of observational unit forms a table.
6. The five most common problems in messy datasets:
- Column headers aren't variable names
- Multiple variables are stored in a single column
- Variables are stored in rows and columns
- Multiple observational unit types in the same table
- A single observational unit is stored in multiple tables
Table 4 is messy because the variables aren't uniformly in columns.
Melting is when you turn columns into rows.
7. Table 11 is messy because it has variables in both rows and columns.
Table 12 is molten because each row represents the meteorological measurements for a single day.
8. The chicken-and-egg problem with tidy data is, tidy data is valuable when there are good tools available that work well with it, but if we only focus on creating tidy data, we might get stuck if those tools can't fully improve upon the data structure. Also, only focusing on improving tools might not make progress if the structures they are supposed to fix are not well organized. This creates a loop where neither the data nor tools improve because they are both dependent on each other.
In the future, Hadley Wickham hopes that there is advancement beyond tidy data to develop even better data storage and organization tools. He also hopes there will be continued effort and innovation in data wrangling, and he acknowledges it is hard to break out of the "local maxima".
